# Data Preprocessing

```python
data = data.drop(["customerID"],axis=1)
```

```python
data.head()
```

```python
le = LabelEncoder()

labels_to_encode = [
    'Churn', 
    'gender', 
    'Partner',
    'Dependents',
    'PhoneService',
    'MultipleLines',
    'InternetService',
    'OnlineSecurity',
    'OnlineBackup',
    'DeviceProtection',
    'TechSupport',
    'StreamingTV',
    'StreamingMovies',
    'Contract',
    'PaperlessBilling',
    'PaymentMethod'
]

for column in labels_to_encode:
    data[column] = le.fit_transform(data[column])
```

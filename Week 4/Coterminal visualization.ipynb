```python
import matplotlib.pyplot as plt
import numpy as np
```

# base angle

```python
theta_base = -7
```

# coterminal angles

```python
theta_positive = theta_base + 2 * np.pi
theta_negative = theta_base - 2 * np.pi
```

# unit circle

```python
circle = np.linspace(0, 2 * np.pi, 300)
```

# angle coordinates

```python
x_base, y_base = np.cos(theta_base), np.sin(theta_base)
x_pos, y_pos = np.cos(theta_positive), np.sin(theta_positive)
x_neg, y_neg = np.cos(theta_negative), np.sin(theta_negative)
```

# plot points

```python
fig, ax = plt.subplots()
ax.plot(np.cos(circle), np.sin(circle), label="Unit Circle", color="lightgray")
ax.plot([0, x_base], [0, y_base], label="-7 rad", color="blue", linewidth=3)
ax.plot([0, x_pos], [0, y_pos], label="-7 + 2π rad", color="green", linewidth=2)
ax.plot([0, x_neg], [0, y_neg], label="-7 - 2π rad", color="red", linestyle="--", linewidth=2)

ax.annotate("", xy=(1.2, 0), xytext=(-1.2, 0), arrowprops=dict(arrowstyle="<->", color='black'))
ax.annotate("", xy=(0, 1.2), xytext=(0, -1.2), arrowprops=dict(arrowstyle="<->", color='black'))

ax.text(1.25, 0, 'x', fontsize=12, va='center')
ax.text(0, 1.25, 'y', fontsize=12, ha='center')

ax.set_aspect('equal')
ax.grid(True, linestyle='--', alpha=0.5)
ax.legend()
ax.set_title("Coterminal Angles on the Unit Circle with X and Y Axes")
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.axis('off') 

plt.show()
```
